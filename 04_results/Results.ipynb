```python
def print_sep(name):
    print("\n"*2)
    print("-"*50)
    print(name)

    
    
def print_results(evaluator, round=2):
    hit_rate = evaluator.get_hit_rates()
    recall = evaluator.get_recalls()
    mrr = evaluator.get_mrr()
    times = evaluator.get_times()
    fit_times = evaluator.get_fit_per_year_times()
    
    print_sep("Hit-Rate")
    display(hit_rate)
    print(hit_rate.round(round).to_markdown())
    
    print_sep("Recall")
    display(recall)
    print(recall.round(round).to_markdown())
    
    print_sep("MRR")
    display(mrr)
    print(mrr.round(round).to_markdown())
    
    print_sep("Times")
    display(times)
    print(times.round(round).to_markdown())
    
    print_sep("Fit times")
    display(fit_times)
    print(fit_times.round(round).to_markdown())
```

## CMF Binary 32 

```python
evaluator.load_results("../results/cmf_binary_32_ranks.csv", "../results/cmf_binary_32_times.csv")

print_results(evaluator)
```
